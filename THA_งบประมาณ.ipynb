```python
import pandas as pd
```

```python
df = pd.read_excel("20231031 ข้อมูล 3.1  ของแผน 131312  กับ 131211  ของปี 66 (1).xlsx",header=1,sheet_name=1)
```

[ปี, แผน, กลุ่มแผนงาน, จำนวนเงินของงบประมาณ, ประเภทของงบประมาณ]

```python
data_pool = []
def transform_data(data):
    year = data[0]
    plan = data[1]
    groupplan = data[2]
    subplan = data[3]
    grant_anumat = ("งปม.ที่อนุมัติให้Grant ได้(บ.)",data[4])
    kang_tor = ("โครงการ อยู่ในกระบวนการก่อนอนุมัติ(ในท่อ)" ,data[5])
    anumat = ("โครงการ อนุมัติ",data[6])
    pr = ("จัดซื้อ PR" ,data[7])
    po = ("จัดซื้อ PO",data[8])
    ror_anumat = ("บริหาร รออนุมัติ",data[9])
    anumat_berk = ("บริหาร อนุมัติเบิกจ่าย",data[10])
    ngob_lue = ("งปม.คงเหลือ(บ.)",data[11])
    ngob_lue_per = ("งปม.คงเหลือ(%)",data[12])
    grant_konglue = ("งปม.คงเหลือ Grant(บ.)", data[13])
    grant_konglue_per = ("งปม.คงเหลือ Grant(%)", data[14])
    
    for n in [grant_anumat, kang_tor, anumat, pr, po, ror_anumat, anumat_berk, ngob_lue, ngob_lue_per, grant_konglue, grant_konglue_per]:
        data_pool.append((year, plan, groupplan, subplan, n[1], n[0]))

df.apply(transform_data, axis=1)
```

```python
final = pd.DataFrame(data_pool, columns=["ปี", "แผน", "กลุ่มแผนงาน", "แผนปฏิบัติการ", "จำนวนเงินของงบประมาณ", "ประเภทของงบประมาณ"])
```
